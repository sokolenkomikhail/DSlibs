### 1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Многое зависит от поставленной задачи. 

Но в целом зависит от цены той или иной ошибки. В случае, когда цена ошибки **первого рода** (ошибочного отнесения объекта к первому классу - FP) **выше** цены ошибки **второго рода** (ошибочного отнесения объекта к нулевому классу - FN), следует максимизировать **_precision_**. В обратном случае следует максимизировать recall.

$$ precision = \frac{TP}{TP + FP} $$

Максимизировать precision следует в тех случаях, когда важно избегать ошибочного отнесения объекта к первому классу, когда он относится к нулевому классу

Пример 1:
Классификация писем (спам - 1, не спам - 0). Во избежание отнесения важного письма к спаму.

Пример 2:
Отбор людей по показателям здоровья для испытания нового препарата. В случае отнесения неподходящих людей (по показателям здоровья) к подходящим, для этих людей будут выше шансы нанесения вреда здоровью.

$$ recall = \frac{TP}{TP + FN} $$

Максимизировать recall следует в тех случаях, когда важно максимально точно классифицировать объекты первого класса. 

Пример 1: 
Отток клиентов. Где склоняющиеся к уходу относятся к первому классу, и напротив - к нулевому. Для минимизации количества клиентов ушедших к конкурентам.

Пример 2:
В кредитном скоринге. Для выявления неблагонадежных заемщиков. Цена выдачи кредита неблагонадежному заемщику выше (возможна потеря всей суммы кредита), чем цена невыдачи кредита благонадежному заемщику (потеря процентов по кредиту).

### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Рассмотрим следующую ситуацию: 

Всего 100 объектов. 90 объектов нулевого класса, 10 объектов первого класса. Алгоритм отнес все объекты к первому классу. Таким образом мы получили recall = 1.0 и precision = 0.1. 

Если брать среднее, то мы получим: 
$$ mean = \frac{precision + recall}{2} = \frac{0.1 + 1.0}{2} = 0.55 $$

Что выглядит не так плохо, но на самом деле результат очень плохой (все объекты отнесены к первому классу).




При использовании F-меры, мы получаем:
$$ F = \frac{2 * precision * recall}{precision + recall} = \frac{2 * 0.1 * 1.0}{0.1 + 1.0} = \frac{2}{11} = 0.(18) $$

Что куда лучше отражает результат
